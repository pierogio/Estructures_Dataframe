### Descripción variables
- ActualElapsedTime: tiempo real transcurrido del vuelo, en minutos (almacenado como número entero).
- AirTime: tiempo en el aire para el vuelo, en minutos (almacenado como número entero).
- ArrDelay: retraso en la llegada, en minutos (almacenado como número entero).
- ArrTime: hora real de llegada (almacenado como número decimal).
- Cancelled: estado de cancelación, 1=Si, 0=No (almacenado como valor lógico).
- CancellationCode: código de cancelación, si corresponde (almacenado como factor).
- CarrierDelay: demora, en minutos, atribuible a la aerolínea (almacenado como número entero).
- CRSArrTime: hora programada de llegada (almacenado como número decimal).
- CRSDepTime: hora programada de salida (almacenado como número decimal).
- CRSElapsedTime: tiempo programado de vuelo, en minutos (almacenado como número entero).
- DayOfMonth: día del mes (1 al 31) (almacenado como número entero).
- DayOfWeek: día de la semana (almacenado como factor).
- DepDelay: retraso en la salida, en minutos (almacenado como número entero).
- DepTime: hora real de salida (almacenado como número decimal).
- Dest: aeropuerto de destino (almacenado como factor).
- Diverted: estado de desvío, 1=Si, 2=No (almacenado como valor lógico).
- Distance: distancia de vuelo (almacenado como número entero).
- FlightNum: número de vuelo (almacenado como factor).
- LateAircraftDelay: retraso, en minutos, atribuible a aeronaves que llegan tarde (almacenado como número entero).
- Month: mes del vuelo (almacenado como factor).
- NASDelay: retraso, en minutos, atribuible al Sistema Nacional de Aviación (almacenado como número entero).
- Origin: aeropuerto de origen (almacenado como factor).
- SecurityDelay: retraso, en minutos, atribuible a factores de seguridad (almacenado como número entero).
- TailNum: número de cola del avión (almacenado como factor).
- TaxiIn: tiempo de taxi desde que aterriza hasta que llega a la puerta de llegada, en minutos (almacenado como número entero).
- TaxiOut: tiempo de taxi desde la salida de la puerta hasta que despega, en minutos (almacenado como número entero).
- UniqueCarrier: identificación de la aerolínea (almacenado como factor).
- WeatherDelay: retraso, en minutos, atribuible a factores climáticos (almacenado como número entero).
- Year: año del vuelo (almacenado como factor).

### Creación de dataframe de 200000 records

```python
#Creación del nuevo df
delayed_sample=delayed_flights.sample(n=200000, random_state=42)
```

### Limpiar Columnas
- Year: como todos los datos en examen son del año 2008, es posible prescindir de esta columna; 
- DepTime, CRSDepTime, ArrTime, CRSArrTime, TaxiIn, TaxiOut, CRSElapsedTime: estas columnas no aportan información relevante para el análisis ya que se puede encontrar informaciones más exhaustivas en referencia a los retrasos en otras columnas;
- CancellationCode: el código de cancelación no nos aporta ninguna información útil.

```python
#Elimino las columnas
delayed_sample=delayed_sample.drop(['Year', 'DepTime', 'CRSDepTime', 'ArrTime', 'CRSArrTime', 'TaxiIn', 'TaxiOut', 'CancellationCode','CRSElapsedTime' ], axis=1)
```

### Identificación de valores que faltan

```python
#Visualización grafica de valores nulos
msno.bar(delayed_sample)
```

```python
#Comprobación de existencia de correlación entre los valores faltantes
msno.heatmap(delayed_sample)
```

```python
#Visualizo los valores nulos
delayed_sample.isnull().sum().sort_values()
```

```python
#Añado valor cero a las celdas NaN para que a la hora de realizar operaciones matemáticas los mismos puedan tener un impacto en los cálculos.
delayed_sample=delayed_sample.fillna(0)
```

```python
delayed_sample.isnull().sum().sort_values()
```

```python
#Elimino las rows que corresponden a las celdas con valor 1 de las columnas "Cancelled" y "Diverted".
#No aportan información y las rows correspondientes tienen valor 0.

delayed_sample = delayed_sample.drop(delayed_sample[delayed_sample['Cancelled']>0].index)
delayed_sample = delayed_sample.drop(delayed_sample[delayed_sample['Diverted']>0].index)
```

### Añado nuevas columnas

```python
#Calculo velocidad media de cada avion y añado nueva columna.
#Me aseguro que no hayan valores inf permitiendo asi que esta columna pueda ser utilizada para hacer calculos como media, max, etc.
delayed_sample['SpeedAverage'] = np.round(delayed_sample['Distance'] / (delayed_sample['AirTime'] / 60), 2).replace([np.inf, -np.inf], 0)
```

```python
#Visualizo las columnas
delayed_sample.columns
```

```python
#Cambio orden columnas
delayed_sample = pd.DataFrame.reindex(delayed_sample, columns = ['Month', 'DayofMonth', 'DayOfWeek', 'UniqueCarrier', 'FlightNum',
       'TailNum', 'ActualElapsedTime', 'AirTime', 'SpeedAverage', 'Distance', 'ArrDelay',
       'DepDelay', 'Origin', 'Dest', 'Cancelled', 'Diverted',
       'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay',
       'LateAircraftDelay'])
```

```python
delayed_sample.head()
```

```python
#Añado columna que indica si el avion ha llegado en horario.
#1 = Yes, 0 = No

delayed_sample['OnTime'] = np.where(delayed_sample['ArrDelay'] <= 0, 1, 0)
```

```python
delayed_sample = pd.DataFrame.reindex(delayed_sample, columns = ['UniqueCarrier', 'FlightNum',
       'TailNum', 'AirTime', 'SpeedAverage', 'Distance', 'OnTime', 'ArrDelay',
       'DepDelay', 'Origin', 'Dest', 'Cancelled', 'Diverted',
       'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay',
       'LateAircraftDelay', 'ActualElapsedTime','Month', 'DayofMonth', 'DayOfWeek', ])
```

### Analisis de los datos

```python
#Datos estadisticos del df
delayed_sample.describe()
#df.replace([np.inf, -np.inf], 0, inplace=True)
```

```python
#Creo un dataframe para ver el porcentaje de delay por causa
delay = np.round(delayed_sample[['SecurityDelay', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'LateAircraftDelay']].sum(),2)
total_delay=np.round(delay.sum(),2)
security_delay_pct = np.round(delayed_sample['SecurityDelay'].sum() / total_delay * 100, 2)
carrier_delay_pct = np.round(delayed_sample['CarrierDelay'].sum() / total_delay * 100, 2)
weather_delay_pct = np.round(delayed_sample['WeatherDelay'].sum() / total_delay * 100, 2)
nas_delay_pct = np.round(delayed_sample['NASDelay'].sum() / total_delay * 100, 2)
late_aircraft_delay_pct = np.round(delayed_sample['LateAircraftDelay'].sum() / total_delay * 100, 2)

delay_tipos = ['Security', 'Carrier', 'Weather', 'NAS', 'Late Aircraft']
delay_porcentajes = [security_delay_pct, carrier_delay_pct, weather_delay_pct, nas_delay_pct, late_aircraft_delay_pct]
delay_final = pd.DataFrame({'Causa Delay': delay_tipos, 'Porcentaje': delay_porcentajes})

delay_final
```

```python
#Visualización grafica del mismo
delay = np.round(delayed_sample[['SecurityDelay', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'LateAircraftDelay']].sum(),2)
total_delay=np.round(delay.sum(),2)
security_delay_pct = np.round(delayed_sample['SecurityDelay'].sum() / total_delay * 100, 2)
carrier_delay_pct = np.round(delayed_sample['CarrierDelay'].sum() / total_delay * 100, 2)
weather_delay_pct = np.round(delayed_sample['WeatherDelay'].sum() / total_delay * 100, 2)
nas_delay_pct = np.round(delayed_sample['NASDelay'].sum() / total_delay * 100, 2)
late_aircraft_delay_pct = np.round(delayed_sample['LateAircraftDelay'].sum() / total_delay * 100, 2)

delay_tipos = ['Security', 'Carrier', 'Weather', 'NAS', 'Late Aircraft']
delay_porcentajes = [security_delay_pct, carrier_delay_pct, weather_delay_pct, nas_delay_pct, late_aircraft_delay_pct]
plt.bar(delay_tipos, delay_porcentajes)

plt.title('Porcentaje del delay por causa')
plt.xlabel('Causa delay')
plt.ylabel('Porcentaje')

plt.xticks(rotation=45)

plt.show()
```

```python
#Visualización grafica aerolinea con mas retraso

company_delay = delayed_sample.loc[:,['UniqueCarrier', 'ArrDelay']]
company_delay = company_delay.groupby(['UniqueCarrier']).sum().reset_index()
company_delay['Percentage'] = np.round((company_delay['ArrDelay']/company_delay['ArrDelay'].sum())*100)
company_delay.rename(columns = {'ArrDelay':'TotalDelay'}, inplace = True)
company_delay = company_delay.sort_values('TotalDelay', ascending=False)
company_delay = company_delay.drop(company_delay[company_delay['Percentage']<=0].index)

plt.bar(company_delay['UniqueCarrier'], company_delay['Percentage'])

plt.title('Delay by Airline')
plt.xlabel('Airline') 
plt.ylabel('Percentage')

plt.xticks(rotation=45)

plt.show()
```
